```python
import matplotlib.pyplot as plt
import json
import numpy as np
from lib.constants import *
```

```python
f = open("output/plot_simple.json")
data = json.load(f)
list(data.keys())
```

```python
labels = ["common_words_distribution" ,"punct_distribution" ,"stop_words_distribution" ,"word_len_distribution" ,"filtered_word_len_distribution" ,"stop_word_len_distribution" ,"tag_distribution"]
temp = [[], [], [], [], [], [], []]
for auth in data.keys():
    for label in data[auth].keys():
        if label not in ['av_words_per_sent', 'av_words_per_para']:
            temp[labels.index(label)].append(data[auth][label])

        
data2 = dict(zip(labels, temp))
len(data2['common_words_distribution'])
```

```python
for label in data2.keys():
    for i in range(len(data2[label])):
        x = range(len(data2[label][i]))
        y = np.cumsum(data2[label][i])
        plt.plot(x, y)
    plt.title(label)
    plt.show()
```
